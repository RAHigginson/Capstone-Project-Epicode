```python
import pandas as pd
```

```python
#Import of the csv file
df = pd.read_csv(r"csv\population.csv")
```

```python
#First look to our population file
df
```

```python
#We look for and delete duplicates
df.drop_duplicates()
```

```python
#We delete the columns we are not going to need
columns_to_drop = ['Other people in need of international protection', 'Stateless persons', 'Host Community','Others of concern']
df.drop(columns=columns_to_drop, inplace=True)
df
```

```python
#check the countries names to identify those with special characters in them 
countries = df['Country of origin'].unique()
print(countries)
```

```python
#deviding the countries with details written among parenthesis and renaming the columns
df[["Country_of_origin", "details_to_delete"]] = df['Country of origin'].str.split("(", n=1, expand=True)
df[["Country_of_asylum", "details_to_delete_asylum"]] = df['Country of asylum'].str.split("(", n=1, expand=True)
```

```python
#swapping the new country of origin column's position with the old one's
column1 = 'Country of origin'
column2 = 'Country_of_origin'
column3 = 'Country of asylum'
column4 = 'Country_of_asylum'


position1 = df.columns.get_loc(column1)
position2 = df.columns.get_loc(column2)
position1 = df.columns.get_loc(column3)
position2 = df.columns.get_loc(column4)


df = df[[
    col if col != column1 and col != column2 else
    column2 if col == column1 else
    column1
    for col in df.columns
]]

df = df[[
    col if col != column3 and col != column4 else
    column4 if col == column3 else
    column3
    for col in df.columns
]]
```

```python
#deleting the last columns
new_columns_to_delete = ["Country of origin", "details_to_delete", 'Country of asylum', "details_to_delete_asylum"]
df.drop(columns=new_columns_to_delete, inplace=True)
```

```python
#renaming the column so it would have its original name
df.rename(columns = {'Country_of_origin' : "Country of origin", 'Country_of_asylum' : 'Country of asylum'}, inplace=True)
df
```

```python
#Trim whitespace from ends of each value across all series in dataframe
def trim_all_columns(df):
    trim_strings = lambda x: x.strip() if isinstance(x, str) else x
    return df.applymap(trim_strings)
df = trim_all_columns(df)
df
```

```python
countries2 = df['Country of origin'].unique()
print(countries2)
```

```python
#deleting rows with country of origin Unknown
unknownfilt = df['Country of origin'] == "Unknown"
df.drop(index=df[unknownfilt].index, inplace=True)
df
```

```python
replace_dict = {
    'Türkiye': 'Turkiye',
    'Serbia and Kosovo: S/RES/1244': 'Serbia and Kosovo',
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
}
df['Country of origin'] = df['Country of origin'].replace(replace_dict)
df['Country of asylum'] = df['Country of asylum'].replace(replace_dict)
```

```python
countries3 = df['Country of origin'].unique()
print(countries3)
```

```python
df.to_csv("population_cleaned.csv", index=False)
```
