```python
#Import of the csv file
df = pd.read_csv(r"population_cleaned.csv")
```

```python
df.sort_values(by="Refugees under UNHCR's mandate", ascending=False).head(10)
```

```python
group_by_country = df.groupby("Country of origin")
```

```python
group_by_country.agg({
    "Refugees under UNHCR's mandate": 'sum',
    "Asylum-seekers": "sum",
    "IDPs of concern to UNHCR": "sum"
}).sort_values(by='Refugees under UNHCR\'s mandate', ascending=False)
```

```python
group_by_country.agg({
    "Refugees under UNHCR's mandate": 'sum',
    "Asylum-seekers": "sum",
    "IDPs of concern to UNHCR": "sum"
}).sort_values(by='Asylum-seekers', ascending=False)
```

```python
group_by_country.agg({
    "Refugees under UNHCR's mandate": 'sum',
    "Asylum-seekers": "sum",
    "IDPs of concern to UNHCR": "sum"
}).sort_values(by='IDPs of concern to UNHCR', ascending=False)
```
